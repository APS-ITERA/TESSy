```python
from astroquery.mast import Observations
from astroquery.mast import Catalogs
from astroquery.mast import Tesscut
from astropy.coordinates import SkyCoord
```

```python
obs_table = Observations.query_criteria(filters='TESS',objectname='WASP_101')
```

```python
print(obs_table)
```

```python
catalog_data = Catalogs.query_object("WASP-101", catalog="TIC", radius=0.2)
print(catalog_data)
```

```python
coord = SkyCoord(98.2915599415135, -23.6785834495921, unit='deg')
sector_table = Tesscut.get_sectors(coordinates=coord)
print(sector_table)
```

```python
sector_table = Tesscut.get_sectors(objectname="WASP-101")
print(sector_table)
```
