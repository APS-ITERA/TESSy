```python
star = eleanor.Source(tic=375506058, sector=15)
```

```python
data = eleanor.TargetData(star, height=15, width=15, bkg_size=31, do_psf=True, do_pca=True, regressors='corner')
```

```python
lc.plot();
```

```python
lk = data.to_lightkurve()
lk.plot()
```
